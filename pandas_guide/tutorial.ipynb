Представим, что в `bios` есть колонки `name`, `email`, `phone`, `bio`.

### 1. Якоря `^` и `$` — привязка к началу/концу строки
```python
# Имена, которые НАЧИНАЮТСЯ с "Dr." (точка экранируется, т.к. в regex . = любой символ)
bios[bios["name"].str.contains(r"^Dr\.", na=False)]

# Биографии, которые ЗАКАНЧИВАЮТСЯ на "PhD"
bios[bios["bio"].str.contains(r"PhD$", na=False)]
```
`^` — начало строки, `$` — конец. Без них `Dr.` нашлось бы и внутри предложения.

### 2. Классы символов и квантификаторы
```python
# Код страны в телефоне: + и 1–3 цифры
bios[bios["phone"].str.contains(r"^\+\d{1,3}\s", na=False)]

# Возраст от 18 до 99 в биографии (две цифры подряд)
bios[bios["bio"].str.contains(r"\b\d{2}\b", na=False)]
```
`\d` — цифра, `{1,3}` — от 1 до 3 повторений, `\b` — граница слова (чтобы "123" не считалось возрастом).

### 3. Валидация email
```python
email_pattern = r"^[\w.+-]+@[\w-]+\.[\w.]+$"
bios[bios["email"].str.contains(email_pattern, na=False)]
```
Разбор: `[\w.+-]+` — один или более символов из букв/цифр/`_`/`.`/`+`/`-`, потом `@`, потом домен, потом точка и TLD.

### 4. Альтернатива с группами — «или» внутри слова
```python
# Имена, содержащие "Ann" или "Anna" (но не "Annabelle")
bios[bios["name"].str.contains(r"\bAnn(a)?\b", na=False)]
```
`(a)?` — необязательная буква `a`, `\b` по краям — чтобы не ловить подстроки.

### 5. Негативный просмотр вперёд — «содержит, НО без ...»
```python
# Все имена, где НЕТ слов "Jr" или "Sr"
bios[bios["name"].str.contains(r"^(?!.*\b(Jr|Sr)\b).*$", na=False)]
```
`(?!...)` — «дальше НЕ должно быть …». Удобно, когда `~str.contains(...)` слишком груб.

### 6. Бонус: `str.extract` — вытащить части в отдельные колонки
Это уже не фильтр, а извлечение — но очень полезно:
```python
# Разделить "Ivanov, Ivan" на фамилию и имя
pattern = r"^(?P<surname>\w+),\s*(?P<first>\w+)$"
bios[["surname", "first"]] = bios["name"].str.extract(pattern)
```
Именованные группы `(?P<name>...)` автоматически становятся названиями колонок.

---

### Пара советов, которые сэкономят нервы:
- **`na=False`** — всегда добавляй в `str.contains`, иначе NaN в колонке даст `NaN` вместо `False`, и фильтрация сломается.
- **`case=False`** — если регистр не важен: `str.contains("anna", case=False)`.
- **`regex=True`** — дефолт в `contains`, но в `str.replace` его нужно явно указывать.
- Используй **raw-строки** `r"..."` — иначе придётся экранировать обратные слеши вдвое: `"\\d+"`.

````python
def categorize_athlete(row):
    """Определение атлета по весу и росту."""
    if row["height_cm"] < 175 and row["weight_kg"] < 70:
        return "Lightweight"
    elif row["height_cm"] < 185 or row["weight_kg"] <= 80:
        return "Middleweight"
    else:
        return "Heavyweight"


bios["Category"] = bios.apply(categorize_athlete, axis=1)  # axis=1 - выбор между строками и столбцами (0 и 1 соответственно)
````
