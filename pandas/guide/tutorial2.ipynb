```python
def increase_salary(x: float) -> float:
    if x < 50:
        return x * 1.1
    if x < 60:
        return x * 1.05
    return x


salaries = pd.Series([60, 45, 50], ["Маша", "Паша", "Женя"])
salaries = salaries.map(increase_salary)
salaries
```

## Задача 1. Зарплаты и премии

Нужно посчитать зарплаты с учётом премий и на сколько в среднем они увеличились. Округлить до 500 рублей

```python
salaries = pd.Series([60, 45, 50], ["Маша", "Паша", "Женя"])
bonuses = pd.Series({"Маша": 2, "Женя": 7})  # % от зарплаты
```

```python
money = salaries * (1 + bonuses / 100)  # Добавляем премии
```

```python
money.fillna(salaries, inplace=True)  # Кому не дали премии, для того берём старое значение
```

```python
money
```

```python
(money - salaries).mean()  # Среднее значение прибавки зарплат
```

```python
money
```

### Решение автора туториала

```python
salaries = pd.Series([60.0, 45.0, 50.0], ["Маша", "Паша", "Женя"])
bonuses = pd.Series({"Маша": 2.0, "Женя": 7.0})  # (% от зарплаты)
```

```python
bonuses = bonuses.reindex(salaries.index).fillna(0)
new_salaries = salaries * (1 + bonuses / 100)
```

```python
new_salaries
```

```python
new_salaries.mean() - salaries.mean()
```
