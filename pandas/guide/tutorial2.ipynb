```python
import pandas as pd
```

```python
salaries = pd.Series([60.0, 45.0, 50.0])  # Колонка зарплат в тысячах рублей
print(salaries)
```

```python
salaries = pd.Series([60.0, 45.0, 50.0], ["Маша", "Паша", "Женя"])  # Добавляем свои индексы
print(salaries)
```

```python
salaries = pd.Series({"Маша": 60.0, "Паша": 45.0, "Женя": 50.0})  # Можно передавать словарь
salaries  # Аналогично print(salaries)
```

```python
print(salaries.values)
```

```python
salaries.values
```

```python
salaries.index  # Аналогично print(salaries.index)
```

```python
salaries.values.tolist()
```

```python
salaries.index.tolist()
```

```python
salaries.to_dict()  # Словарь, где ключи - индексы
```

```python
print(salaries.loc["Женя"])  # Зарпалата Жени (можно без print, но будет больше служебной информации)
```

```python
salaries.loc["Паша"] = 55.0  # Изменяем значение
salaries.loc["Саша"] = 65.0  # Добавляем новую запись
salaries
```

```python
salaries.loc["Маша":"Паша"]  # В отличие от обычных срезов, конечный элемент включается в выборку
```

```python
salaries.loc["Паша":"Женя"] = [55.0, 60.0]  #  Можем менять сразу несколько значений
salaries
```

```python
for name, salary in salaries.items():
    print(f"{name} получает {salary} тысяч рублей.")
```

```python
salaries + 5  # Увеличиваем все зарплаты на 5 тысяч рублей
```

```python
salaries = pd.Series([60.0, 45.0, 50.0], ["Маша", "Паша", "Женя"])
bonuses = pd.Series([5.0, 1.0, 3.0], ["Паша", "Маша", "Женя"])
salaries + bonuses  # Добавим первональные премии к зарплатам
```

```python
# Если явно не указать индексы, они автоматические проставятся и операция останется корректной
a = pd.Series([1, 2, 3])
b = pd.Series([4, 5, 6])
a + b
```

```python
c = (a + b) ** 2  # Если результат нужно сохранить, присваиваем ему переменную
c
```

```python
a *= b  # Оператор составного присвоения тоже работает
a
```

```python
salaries > 55
```

```python
bonuses_2023 = pd.Series([5.0, 1.0, 3.0], ["Паша", "Маша", "Женя"])
bonuses_2024 = pd.Series([7.0, 0.5, 4.0], ["Паша", "Маша", "Женя"])
bonuses_2024 > bonuses_2023
```

```python
(bonuses_2024 > bonuses_2023) & (bonuses_2024 > 1)  # Комбинируем условия
```

```python
index = [True, False, True]
salaries[index]  # Булевская индексация
```

```python
# index = salaries < 55
# salaries[index]

salaries[salaries < 55]  # Лучше одной строкой
```

```python
salaries[salaries < 55] += 5.0
salaries
```

```python
# Если индексы не совпадают, то появляется NaN
a = pd.Series([1, 2, 3], ["A", "B", "C"])
b = pd.Series([4, 5, 6], ["A", "B", "X"])
c = a + b
c
```

```python
# Если длины разные, видим похожую картину
a = pd.Series([1, 2, 3], ["A", "B", "C"])
b = pd.Series([4, 5], ["A", "B"])
c = a + b
c
```

```python
c.loc["C"] is None  # NaN != None
```

```python
c.isna()
```

```python
c.notna()

# Можем избавиться от нанов с помощью булевской индексации
# c = c[c.notna()]
# c
```

```python
c[c.isna()] = 0  # Заполняем наны нулями
c
```
