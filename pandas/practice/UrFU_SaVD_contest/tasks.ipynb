## Задача 1

Напишите функцию, которая создает объект Series из списка натуральных чисел и пользовательскими индексами, состоящими из всех букв латинского алфавита, используя библиотеку pandas.

Функция не должна принимать аргументы. Результат операции выведите с помощью return.

```python
from string import ascii_lowercase


def numbers_series():
    data = pd.Series()
    for numb, symb in enumerate(ascii_lowercase):
        data[symb] = numb + 1
    return data


print(numbers_series())
```

## Задача 4

Используя библиотеку pandas, напишите скрипт, который

- загружает файл SQLite с названием european_database.sqlite
- на его основе создает объект Dataframe
- оставляет в датафрейме одну колонку с удалением дубликатов - 'country'
- сохраняет полученный датафрейм в новый файл с именем new_database.sqlite

Источник данных - https://www.kaggle.com/datasets/groleo/european-football-database

```python
import sqlite3
```

```python
database = pd.read_sql_query("SELECT * FROM divisions", con=sqlite3.connect("data/european_database.sqlite"))
```

```python
df_unique = database[["country"]].drop_duplicates()
```

```python
df_unique
```

```python
df_unique.to_sql("countries", con=sqlite3.connect("data/new_database.sqlite"), index=False, if_exists="replace")  # если уже существует - перезаписать
```

## Задача 5

Используя библиотеку pandas, напишите скрипт, который

- загружает файл JSON с названием games_metadata.json
- на его основе создает объект Dataframe
- выполняет фильтрацию игр с тегами Strategy и RTS
- создает сводную таблицу с подсчетом количества игр по жанрам
- сортирует по количеству игр
- сохраняет полученный датафрейм в новый файл с именем games_strategy_rts.json

Источник данных - https://www.kaggle.com/datasets/antonkozyriev/game-recommendations-on-steam

```python
games_metadata = pd.read_json("data/games_metadata.json", orient="records", lines=True)
```

```python
games_metadata.head()
```

```python
games_metadata.info()
```

```python
games_metadata["tags"]
```

```python
games_metadata = games_metadata[games_metadata['tags'].apply(lambda tags: "Strategy" and "RTS" in tags)]
```

```python
games_metadata
```

```python
results = games_metadata.explode("tags")
results
```

```python
pivot = games_metadata.explode("tags").pivot_table(index="tags", values="app_id", aggfunc="count")
```

```python
pivot
```

```python
pivot.sort_values("app_id", inplace=True)
```

```python
pivot
```

```python
pivot.to_json("data/games_strategy_rts.json")
```
