## Задача 1

Дан файл "titanic.csv" в формате .csv.

При проверке файл будет находиться в одной директории с Вашим решением, поэтому его можно открыть командой pd.read_csv("titanic.csv").

Каждая строка этого набора данных - это отдельный человек, которому присвоен набор свойств - имя, номер каюты, возраст и т.д.

В данной задаче необходимо определить количество людей в наборе данных, возраст которых находится в диапазоне строго между M и N, где M и N - целые числа от 0 до 100, которые подаются на вход Вашего скрипта в виде аргументов. Решение нужно оформить в виде функции, принимающей два аргумента (например, ageSearch(M, N)). Функция должна возвращать одно целое число - количество пассажиров Титаника, удовлетворяющих условию.

```python
titanic = pd.read_csv("data/titanic.csv")
```

```python
titanic.head()
```

```python
def ageSearch(M, N):
    titanic_forage = titanic[(titanic["Age"] > M) & (titanic["Age"] < N)]
    return len(titanic_forage)


ageSearch(1, 10)
```

## Задача 10

Добавьте в таблицу столбец с отношением городского населения к общему населению для каждой страны.

Создайте новый столбец "Urban Population Ratio", который будет содержать отношение городского населения (Urban_population) к общему населению (Population) для каждой страны.

Убедитесь, что значения в новом столбце имеют числовой формат (float).

Сохраните таблицу с этим новым столбцом в файл urban_population_ratio.csv.

К файлу-источнику можно обратиться по имени world-data-2023.csv.

Источник данных - https://www.kaggle.com/datasets/nelgiriyewithana/countries-of-the-world-2023

```python
countries = pd.read_csv("data/world-data-2023.csv")
```

```python
countries["Urban_population"]
```

```python
countries["Population"]
```

```python
countries["Urban_population"] = countries["Urban_population"].str.replace(",", "")
countries["Population"] = countries["Population"].str.replace(",", "")
```

```python
countries["Urban_population"] = pd.to_numeric(countries["Urban_population"], errors="coerce")
countries["Population"] = pd.to_numeric(countries["Population"], errors="coerce")
```

```python
countries["Urban Population Ratio"] = countries["Urban_population"] / countries["Population"]
```

```python
countries
```

```python
countries.to_csv("data/urban_population_ratio.csv", index=False)
```
