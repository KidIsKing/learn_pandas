# 5. Работа с файлами XLSX в pandas

Используя библиотеку pandas, напишите скрипт, который

- загружает файл XLSX с названием 54-iip_25.xlsx
- удаляет заголовки и футер (в том числе с датами, а также пустые строки), в качестве индексов принимает первый столбец
- на его основе создает объект Dataframe
- из столбцов данных сохраняет только данные на 01.01.2025
- сохраняет полученный датафрейм в новый файл с именем processed_data.xlsx с сохранением индексов

Источник данных - https://www.cbr.ru/vfs/statistics/credit_statistics/iip/54-iip_25.xlsx

Ссылка на задание: https://elearn.urfu.ru/mod/book/view.php?id=287538&chapterid=18498#mod_book-chapter

---

С ссылкой на файл возникла проблема. Ниже код, который писала в семестре (сейчас лето).

```python
import pandas as pd
import requests

# Скачиваем файл с сайта ЦБ РФ
url = "https://www.cbr.ru/vfs/statistics/credit_statistics/iip/54-iip_25.xlsx"
response = requests.get(url)
with open("54-iip_25.xlsx", "wb") as f:
    f.write(response.content)

# Загружаем Excel, первый столбец как индекс
df = pd.read_excel("54-iip_25.xlsx", index_col=0)

# Удаляем полностью пустые строки (убираем футер и мусор)
df = df.dropna(how='all')

# Оставляем только столбцы, содержащие дату 01.01.2025
# Приводим названия колонок к строке для безопасного поиска
df = df.loc[:, df.columns.astype(str).str.contains('01.01.2025', na=False)]

# Сохраняем результат с сохранением индексов
df.to_excel("processed_data.xlsx", index=True)
```
