### Поподробнее про random_state=

```python
import pandas as pd


df = pd.DataFrame({'A': [1, 2, 3, 4, 5], 'B': [6, 7, 8, 9, 10]})

# Без random_state — результат каждый раз разный
print(df.sample(n=2))

# С random_state=42 — результат всегда одинаковый
print(df.sample(n=2, random_state=42))
```

```python
coffee.loc[[]]  # Шапка таблицы
coffee.loc[[1]]  # Шапка таблицы + 1 запись
coffee.loc[[1, 4, 8]]  # Шапка таблицы + конкретные записи

# Срезы
coffee.loc[7:]
coffee.loc[:7]
coffee.loc[5:7]

# Можно выбирать конкретные строки кроме среза для записей
coffee.loc[5:7, "Coffee Type"]
coffee.loc[5:7, ["Day", "Coffee Type"]]
coffee.loc[:, ["Day", "Coffee Type"]]  # Все записи, но конкретные столбцы

coffee.iloc[:, [0, 1]]  # В .iloc[] для выборки столбцов пишем индексы, а не имена
```

```python
coffee.index = coffee["Day"]  # Сами задаём (переопределяем) индексы для фрейма
coffee.head()

coffee.loc["Monday":"Friday", "Coffee Type"]  # Теперь пользуемся именнованными индексами, в противном случае будет ошибка
coffee.iloc[0:2]  # Тут всё как и было, только в таблицы числовых индексов больше нет
```
