# 1. Работа с файлами CSV в pandas

Используя библиотеку pandas, напишите скрипт, в котором

- Загрузить файл CSV с названием vgsales.csv
- На его основе создать объект Dataframe
- Привести Year к числовому типу
- Удалить строки с пропущенным Year
- Проверить типы столбцов продаж (float)
- Создать столбец Total_Regional_Sales = NA + EU + JP + Other
- Создать столбец Sales_Diff = Global_Sales - Total_Regional_Sales
- Сформировать итоговый DataFrame с колонками:
    - Name, Platform, Year, Genre, Publisher, Global_Sales, Total_Regional_Sales, Sales_Diff
- Выбрать строки, где Sales_Diff != 0
- Отсортировать по убыванию Sales_Diff
- Сохранить полученный датафрейм в новый файл с именем vgsales_cleaned_analysis.csv без сохранения индексов

Файл CSV представлен по ссылке - https://www.kaggle.com/datasets/gregorut/videogamesales

```python
import pandas as pd
```

```python
vgsales = pd.read_csv("data/task1/vgsales.csv")  # Читаем и создаём фрейм данных из файла vgsales.csv
```

```python
vgsales.info()  # Проверяем типы данных
```

```python
vgsales[vgsales["Year"].isna()]  # Вывод строк, где в столбце "Year" пусто
```

```python
vgsales["Year"] = pd.to_numeric(vgsales["Year"])  # Приводим столбец "Year" к числовому типу данных
```

```python
vgsales.info()  # Проверяем типы данных
```

```python
vgsales.dropna(subset=["Year"], inplace=True)  # Удаляем строки, где в столбце "Year" пусто
```

```python
vgsales[vgsales["Year"].isna()]  # Проверяем, что строки с пустыми значениями в столбце "Year" удалены
```

```python
vgsales["Total_Regional_Sales"] = vgsales["NA_Sales"] + vgsales["EU_Sales"] + vgsales["JP_Sales"] + vgsales["Other_Sales"]  # Создаём новый столбец "Total_Regional_Sales"
```

```python
vgsales[abs(vgsales["Global_Sales"] - vgsales["Total_Regional_Sales"]) > 1]  # Выводим строки, где разница между "Global_Sales" и "Total_Regional_Sales" больше 1
```

```python
vgsales["Sales_Diff"] = vgsales["Global_Sales"] - vgsales["Total_Regional_Sales"]  # Создаём новый столбец "Sales_Diff"
```

```python
vgsales.head()
```

```python
vgsales = vgsales[["Name", "Platform", "Year", "Genre", "Publisher", "Global_Sales", "Total_Regional_Sales", "Sales_Diff"]]
```

```python
vgsales
```

```python
vgsales = vgsales[vgsales["Sales_Diff"] != 0]  # Удаляем строки, где разница между "Global_Sales" и "Total_Regional_Sales" равна 0
```

```python
vgsales[vgsales["Sales_Diff"] == 0]  # Проверяем, что строки с разницей между "Global_Sales" и "Total_Regional_Sales", равной 0, удалены
```

```python
vgsales.sort_values("Sales_Diff", ascending=False, inplace=True)  # Сортируем по столбцу "Sales_Diff" в порядке убывания
```

```python
vgsales.to_csv("data/task1/vgsales_cleaned_analysis.csv", index=False)  # Сохраняем очищенный и обработанный фрейм данных в новый CSV файл
```
