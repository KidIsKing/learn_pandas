# 2. Работа с SQLite в pandas

Используя библиотеку pandas, напишите скрипт, который

- Загружает файл SQLite с названием database.sqlite
- На его основе создает объект Dataframe
- Выполняет 3 действия с датафреймом:
    - оставляет только колонку airline
    - удаляет дубликаты значений
    - сортирует значения по алфавит
- Сохраняет полученный датафрейм в новый файл с именем airline.sqlite

Источник данных - https://www.kaggle.com/datasets/crowdflower/twitter-airline-sentiment

```python
import pandas as pd
```

```python
import sqlite3
```

```python
database = pd.read_sql_query("SELECT * FROM Tweets", con=sqlite3.connect("data/task2/database.sqlite"))
```

```python
database.head()
```

```python
database = database[["airline"]]
```

```python
database
```

```python
database.drop_duplicates(inplace=True)  # Удаляем дубликаты
```

```python
database
```

```python
database.sort_values("airline", inplace=True)
```

```python
database
```

```python
database.to_sql("airline", con=sqlite3.connect("data/task2/airline.sqlite"))
```
